```python
import pandas as pd
import numpy as np
import os
```

```python
ids = []
for path, subdirs, files in os.walk("./reviews"):
    for filename in files:
        id = str(filename).split(".")[0]
        ids.append(id)
        
print(ids)
```

```python
d = {"movie_ids": ids}
df = pd.DataFrame(data=d)

# df.head()
df.to_csv("movie_ids.csv")
```
